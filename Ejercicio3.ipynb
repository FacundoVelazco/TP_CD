```python
import pandas as pd
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv',sep=';')
```

```python
jurisdicciones = df_alumnos.iloc[:,0]
```

```python
unique_values = jurisdicciones.unique()
```

```python
diff_jurisd_map = {}
for item in unique_values:
    diff_jurisd_map[str(item)] = jurisdicciones.loc[jurisdicciones == item].size
```

```python
df_final_values = pd.DataFrame(diff_jurisd_map.items())
df_final_values.columns = ['Provincia','Cant. Alumnos']
df_final_values = df_final_values.sort_values(by=['Cant. Alumnos'],ascending=False)
df_final_values = df_final_values.iloc[0:10]   
```

```python
df_final_values = df_final_values.replace(
    ['6','14','82','50','2','90','38','86','70','66'],
    ['Buenos Aires','Córdoba','Santa Fe','Mendoza','CABA','Tucumán','Jujuy','San. del Estero','San Juan','Salta'])
```

```python
ax = df_final_values.plot.bar(x='Provincia',y='Cant. Alumnos',rot=45)
```
