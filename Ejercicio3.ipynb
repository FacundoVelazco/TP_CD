```python
import pandas as pd
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv',sep=';')
```

```python
jurisdicciones = pd.DataFrame(df_alumnos.iloc[:,0])
```

```python
jurisdicciones = jurisdicciones.groupby(['jurisdiccion'])[['jurisdiccion']].count()
```

```python
jurisdicciones.columns = ['Cant. Alumnos']
jurisdicciones.reset_index(inplace=True)
```

```python
jurisdicciones = jurisdicciones.sort_values(by=['Cant. Alumnos'],ascending=False)
jurisdicciones = jurisdicciones.iloc[0:10]   
```

```python
jurisdicciones = jurisdicciones.replace(
    [6,14,82,50,2,90,38,86,70,66],
    ['Buenos Aires','Córdoba','Santa Fe','Mendoza','CABA','Tucumán','Jujuy','San. del Estero','San Juan','Salta'])
```

```python
ax = jurisdicciones.plot.bar(x='jurisdiccion',y='Cant. Alumnos',rot=55)
```
