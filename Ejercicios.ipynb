```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv', sep=';')
df1 = df.iloc[4061:, 29]
df1.replace(-9, np.nan)
df1.replace(-8, np.nan)
df1.replace(-6, np.nan)
df2 = df1.dropna()

fig, ax = plt.subplots()
siTienen = df2.value_counts()[1]
noTienen = df2.value_counts()[2]
ax.pie([siTienen, noTienen], labels=["Tienen acceso a internet", "No tienen acceso a internet"], autopct='%1.1f%%',
       shadow=True)
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv', sep=';')
ages = df_alumnos.iloc[:, 111]

ages = ages.replace(-9, np.nan)
ages = ages.replace('-9', np.nan)
ages = ages.replace(' ', np.nan)

ages = ages.dropna()
ages = ages.replace('1', 1)
ages = ages.replace('2', 2)
ages = ages.replace('3', 3)
ages = ages.replace('4', 4)
ages = ages.replace('5', 5)
pop_size = ages.size

unique_values = ages.unique()
sub_sets_size = {}
for item in unique_values:
       sub_sets_size[str(item)] = ages.loc[ages == item].size
pop_under_13 = pop_size - sub_sets_size['5']
pop_above_13 = sub_sets_size['5']
fig, ax = plt.subplots()
ax.pie([pop_under_13, pop_above_13], labels=["Estudiantes de 13 años o menores", "Estudiantes mayores de 13"],
       autopct='%1.1f%%', shadow=True)
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv', sep=';')
jurisdicciones = pd.DataFrame(df_alumnos.iloc[:, 0])
jurisdicciones = jurisdicciones.groupby(['jurisdiccion'])[['jurisdiccion']].count()
jurisdicciones.columns = ['Cant. Alumnos']
jurisdicciones.reset_index(inplace=True)
jurisdicciones = jurisdicciones.sort_values(by=['Cant. Alumnos'], ascending=False)
jurisdicciones = jurisdicciones.iloc[0:10]
jurisdicciones = jurisdicciones.replace(
       [6, 14, 82, 50, 2, 90, 38, 86, 70, 66],
       ['Buenos Aires', 'Córdoba', 'Santa Fe', 'Mendoza', 'CABA', 'Tucumán', 'Jujuy', 'San. del Estero', 'San Juan',
        'Salta'])
ax = jurisdicciones.plot.bar(x='jurisdiccion', y='Cant. Alumnos', rot=55)
```

```python
df = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv', sep=';')
df1 = df.iloc[:, [5, 106]]

condition = lambda x: x != '-9'
condition2 = lambda x: x != -9
condition3 = lambda x: x != '-6'
condition4 = lambda x: x != -6
condition5 = lambda x: x != ' '
df1 = df1[condition]
df1 = df1[condition2]
df1 = df1[condition3]
df1 = df1[condition4]
df1 = df1[condition5]
cuenta = df1.groupby(['ambito', 'mdesemp'])['ambito'].count()
categorias = ['Debajo del \n básico', 'Básico', 'Satisfactorio', 'Avanzado']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle('Resultados en matemáticas según ambito.', fontsize=12)
ax1.bar(categorias, cuenta[1])
ax1.set_xlabel('Categorias')
ax1.set_ylabel('Resultados')
ax1.set_title('Resultados ambito rural')

ax2.bar(categorias, cuenta[2])
ax2.set_xlabel('Categorias')
ax2.set_ylabel('Resultados')
ax2.set_title('Resultados ambito urbano')

ax1.legend()
ax2.legend()

plt.show()
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv', sep=';')
df_alumnos = df_alumnos.iloc[:, [0, 105, 106]]
df_alumnos_sf = df_alumnos.loc[df_alumnos.jurisdiccion == 82]
df_Ldesemp_sf = pd.DataFrame(df_alumnos_sf['ldesemp'])
df_Mdesemp_sf = pd.DataFrame(df_alumnos_sf['mdesemp'])
df_Ldesemp_sf.reset_index(inplace=True)
df_Ldesemp_sf.drop(['index'], axis=1, inplace=True)
df_Mdesemp_sf.reset_index(inplace=True)
df_Mdesemp_sf.drop(['index'], axis=1, inplace=True)
df_Ldesemp_sf = df_Ldesemp_sf.replace(' ', np.NaN)
df_Ldesemp_sf.dropna()
df_Mdesemp_sf = df_Mdesemp_sf.replace(' ', np.NaN)
df_Mdesemp_sf.dropna()
df_Ldesemp_sf_gb = pd.DataFrame(df_Ldesemp_sf.groupby(['ldesemp'])['ldesemp'].count())
df_Mdesemp_sf_gb = pd.DataFrame(df_Mdesemp_sf.groupby(['mdesemp'])['mdesemp'].count())
df_Ldesemp_sf_gb.columns = ['Cantidad']
df_Mdesemp_sf_gb.columns = ['Cantidad']
df_Ldesemp_sf_gb.reset_index(inplace=True)
df_Mdesemp_sf_gb.reset_index(inplace=True)
df_Ldesemp_sf_gb = df_Ldesemp_sf_gb.replace(
       ['1', '2', '3', '4'],
       ['Debajo del básico', 'Básico', 'Satisfactorio', 'Avanzado'])
df_Mdesemp_sf_gb = df_Mdesemp_sf_gb.replace(
       ['1', '2', '3', '4'],
       ['Debajo del básico', 'Básico', 'Satisfactorio', 'Avanzado'])
cuenta = [
       [df_Ldesemp_sf_gb.iloc[0, 1],
        df_Ldesemp_sf_gb.iloc[1, 1] + df_Ldesemp_sf_gb.iloc[2, 1] + df_Ldesemp_sf_gb.iloc[3, 1]],
       [df_Mdesemp_sf_gb.iloc[0, 1],
        df_Mdesemp_sf_gb.iloc[1, 1] + df_Mdesemp_sf_gb.iloc[2, 1] + df_Mdesemp_sf_gb.iloc[3, 1]]
]
categorias = ['Debajo del \n básico', 'Encima del básico']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle('Proporcion de alumnos por debajo y por encima del nivel baásico en Matematica y Lengua ', fontsize=12)
ax1.bar(categorias, cuenta[0])
ax1.set_xlabel('Categorias')
ax1.set_ylabel('Resultados')
ax1.set_title('Lengua')

ax2.bar(categorias, cuenta[1])
ax2.set_xlabel('Categorias')
ax2.set_ylabel('Resultados')
ax2.set_title('Matemática')

ax1.legend()
ax2.legend()

plt.show()
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv', sep=';')
seleccion = df_alumnos[df_alumnos['jurisdiccion'] == 82]
seleccion['ambito'] = np.where(seleccion['ambito'] == 1, True, False)
seleccion['edadA_junio2022'] = pd.to_numeric(seleccion['edadA_junio2022'], errors='coerce')
seleccion['edadA_junio2022'] = ((seleccion['edadA_junio2022'] != 1) & (seleccion['edadA_junio2022'] != -9))
nuevo_data = pd.DataFrame({
    'indice': range(len(seleccion)),
    'dmat': seleccion['mdesemp'],
    'dlen': seleccion['ldesemp'],
    'arural': seleccion['ambito'],
    'mayor12': seleccion['edadA_junio2022']
})
asignaciones = {"1": "BAJO", "2": "BASICO", "3": "SATISFACTORIO", "4": "AVANZADO"}
nuevo_data['dmat'] = nuevo_data['dmat'].map(asignaciones)
nuevo_data['dlen'] = nuevo_data['dlen'].map(asignaciones)
nuevo_data
nuevo_data.to_json('Data/ej_8.json', orient='records')
```
