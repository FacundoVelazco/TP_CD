```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv',sep=';')
```

```python
df_ambito = df_alumnos.groupby(['ambito'])['ambito'].count()
fig,ax = plt.subplots()
rurales = df_ambito[1]
urbanos = df_ambito[2]
ax.pie([urbanos,rurales],labels=["Urbano","Rural"],autopct='%1.1f%%',shadow=True)
```
