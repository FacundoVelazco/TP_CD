```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv',sep=';')
```

```python
df1 =df.iloc[:,[5,106]]

condition = lambda x: x != '-9'
condition2 = lambda x: x != -9
condition3 = lambda x: x != '-6'
condition4 = lambda x: x != -6
condition5 = lambda x: x != ' '
df1 = df1[condition]
df1 = df1[condition2]
df1 = df1[condition3]
df1 = df1[condition4]
df1 = df1[condition5]
```

```python
cuenta = df1.groupby(['ambito','mdesemp'])['ambito'].count()
```

```python
categorias = ['Debajo del \n básico','Básico','Satisfactorio','Avanzado']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle('Resultados en matemáticas según ambito.', fontsize=12 )
ax1.bar(categorias, cuenta[1])
ax1.set_xlabel('Categorias')
ax1.set_ylabel('Resultados')
ax1.set_title('Resultados ambito rural')

ax2.bar(categorias, cuenta[2])
ax2.set_xlabel('Categorias')
ax2.set_ylabel('Resultados')
ax2.set_title('Resultados ambito urbano')

ax1.legend()
ax2.legend()

plt.show()
```
