```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv',sep=';')

df1 = df.groupby(['ambito'])['ambito'].count()
df1 = df1.sort_values(ascending=False)
print(df1)
fig,ax = plt.subplots()
urbanos = df1[1]
rurales = df1[2]
ax.pie([urbanos,rurales],labels=["Urbano","Rural"],autopct='%1.1f%%',shadow=True)
```
