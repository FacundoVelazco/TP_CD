```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_alumnos = pd.read_csv('Data/Base Aprender estudiantes 6 primaria 2022 – Muestral L y M.csv',sep=';')
```

```python
seleccion = df_alumnos[df_alumnos['jurisdiccion'] == 82][["jurisdiccion", "lpuntaje", "mpuntaje"]]
seleccion
```

```python
seleccion['lpuntaje'] = pd.to_numeric(seleccion['lpuntaje'].str.replace(',', '.'), errors='coerce')
seleccion['mpuntaje'] = pd.to_numeric(seleccion['mpuntaje'].str.replace(',', '.'), errors='coerce')

seleccion = seleccion.dropna(subset=['lpuntaje', 'mpuntaje'])
```

```python
# Calcular el promedio de las columnas "lpuntaje" y "mpuntaje"
promedio_lpuntaje = seleccion['lpuntaje'].mean()
promedio_mpuntaje = seleccion['mpuntaje'].mean()

print("Promedio de lpuntaje:", promedio_lpuntaje)
print("Promedio de mpuntaje:", promedio_mpuntaje)
```

```python
seleccion.reset_index(inplace=True)
seleccion.drop(['index','jurisdiccion'],axis=1,inplace=True)
```

```python
seleccion.columns = ['Puntaje lengua','Puntaje matematica']
seleccion.plot.density()
```
